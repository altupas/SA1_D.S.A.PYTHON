Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average.

```python
import random
import time
import matplotlib.pyplot as plt

def test_sorting_time(sizes):
    times = []

    for size in sizes:

        data = [random.randint(0, 100000) for _ in range(size)]

        start_time = time.time()
        sorted_data = sorted(data)
        end_time = time.time()

        elapsed_time = end_time - start_time
        times.append(elapsed_time)

        print(f"Sorted {size} elements in {elapsed_time:.6f} seconds.")

    return times

dataset_sizes = [10**2, 10**3, 10**4, 10**5, 10**6, 10**7]
sorting_times = test_sorting_time(dataset_sizes)

plt.figure(figsize=(10, 6))
plt.plot(dataset_sizes, sorting_times, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.title("Sorting Time Analysis of Python's sorted() Method")
plt.xlabel("Dataset Size (n)")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()
```
